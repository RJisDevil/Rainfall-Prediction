### Outliers Capping

```python
# outliers capping from the num_df
Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

num_df = num_df.mask(num_df < lower_bound, lower_bound, axis=1)
num_df = num_df.mask(num_df > upper_bound, upper_bound, axis=1)
```

### Train-Test split

```python
# Train test split
X = new_df.drop('RainTomorrow' , axis = 1)
Y = new_df['RainTomorrow']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
# Checking training data of target variable is balanced or not
sns.countplot(x=Y_train, data=new_df , palette = 'viridis')
```

#### To make the dataset balanced SMOTE-NC method is used , it is one of oversampling technique.

```python
# smote-nc sampling for train data
smote_nc = SMOTENC(categorical_features=[0,1,2,3,4,5,6,7,8], random_state=42) # categorical features index from new_df
X_train_res, Y_train_res = smote_nc.fit_resample(X_train, Y_train)

print("X_train_res shape:", X_train_res.shape)
print("Y_train_res shape:", Y_train_res.shape)
```

```python
# Checking resampled training data of target variable is balanced or not
sns.countplot(x=Y_train_res, data=new_df)
```

```python
# Creating function for different classification metrics
def classification_metrics(Y_test, Y_pred, proba=None):
    
    # Compute accuracy score
    acc = accuracy_score(Y_test, Y_pred)
    print(f"Accuracy score: {acc:.3f}\n")
    
    # Compute confusion matrix
    cm = confusion_matrix(Y_test, Y_pred)
    print("Confusion matrix:")
    print(cm)
    
    # Plot confusion matrix
    sns.heatmap(cm, annot=True, cmap="Blues", fmt="d")
    plt.title("Confusion matrix")
    plt.xlabel("Predicted label")
    plt.ylabel("True label")
    plt.show()
    
    # Compute classification report
    report = classification_report(Y_test, Y_pred)
    print("Classification report:")
    print(report)
      
    # proba = classifier.predict_proba(X_test)

    # for binary
    if proba is not None:
        fpr, tpr, thresholds = roc_curve(Y_test, proba[:, 1])
        roc_auc = auc(fpr, tpr)
        
        # Plot ROC curve
        plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
        plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
        plt.title("ROC curve")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.legend(loc = 'lower right')
        plt.show()
        
        # Compute and print AUC score
        print(f"AUC score: {roc_auc:.3f}\n")
          
    # return values
    return acc, cm[0][0],cm[0][1], cm[1][0],cm[1][1], roc_auc if proba is not None else None
```

## Data training with different models

```python
# Creating empty dictionary to store the classification metrics different scores
results = {}
```

### Logistic Regression

```python
# Appling logistic regression on resampled train and test data
logreg = LogisticRegression(C = 10 , max_iter = 100 , penalty = 'l2' , solver = 'liblinear' , random_state=42)
logreg.fit(X_train_res, Y_train_res)
Y_pred_lr = logreg.predict(X_test)
```

```python
# Predicting probabilities and printing classification metrics
proba_lr = logreg.predict_proba(X_test)
results['LogisticRegression']=classification_metrics(Y_test, Y_pred_lr , proba = proba_lr)
```

### Decision Tree

```python
# Appling decision tree on resampled train and test data
dtc = DecisionTreeClassifier(max_depth = 20, max_features = 'log2')
dtc.fit(X_train_res, Y_train_res)
Y_pred_dt = dtc.predict(X_test)
```

```python
# Predicting probabilities and printing classification metrics
proba_dt = dtc.predict_proba(X_test)
results['DecisionTree']=classification_metrics(Y_test, Y_pred_dt, proba=proba_dt)
```

### Random Forest

```python
# Appling random forest on resampled train and test data
rfc = RandomForestClassifier()
rfc.fit(X_train_res, Y_train_res)
Y_pred_rf = rfc.predict(X_test)
```

```python
# Predicting probabilities and printing classification metrics
proba_rf = rfc.predict_proba(X_test)
results['RandomForest']=classification_metrics(Y_test, Y_pred_rf, proba=proba_rf)
```

### XGBoost (Extreme Gradient Boosting)

```python
# Appling xgboost on resampled train and test data
xgbc = XGBClassifier(max_depth = 20, n_estimators = 125, random_state = 50)
xgbc.fit(X_train_res, Y_train_res)
Y_pred_xgb = xgbc.predict(X_test)
```

```python
# Predicting probabilities and printing classification metrics
proba_xgb = xgbc.predict_proba(X_test)
results['XGBoost']=classification_metrics(Y_test, Y_pred_xgb, proba=proba_xgb)
```

### AdaBoost (Adaptive Boosting)

```python
# Appling adaboost on resampled train and test data
adabc = AdaBoostClassifier(n_estimators = 150, random_state = 25)
adabc.fit(X_train_res, Y_train_res)
Y_pred_adab = adabc.predict(X_test)
```

```python
# Predicting probabilities and printing classification metrics
proba_adab = adabc.predict_proba(X_test)
results['AdaBoost']=classification_metrics(Y_test, Y_pred_adab, proba=proba_adab)
```

```python
# Creating a pandas dataframe from the results dictionary
results_df = pd.DataFrame.from_dict(results, orient="index")
results_df=results_df.rename(columns={0 :'Accuracy',1:'TN',2:'FP',3:'FN',4:'TP' ,5 : 'AUC'})
```

```python
# Adding columns precision , recall , f-1 score 
results_df['Precision']=results_df['TP']/(results_df['TP']+results_df['FP'])
results_df['Recall']=results_df['TP']/(results_df['TP']+results_df['FN'])
results_df['f-1_score'] = 2*results_df['Precision']*results_df['Recall']/(results_df['Precision']+results_df['Recall'])
```
